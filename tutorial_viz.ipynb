```r
#goal of this assignment is to create one plot that can help you figure out which restaurant to open and where
library(tidyverse)

fast_food <- read_csv("data/fast_food.csv")
top_restaurants <- fast_food |>
    group_by(name) |>
    summarise(n = n()) |>
    arrange(desc(n)) |> 
    head(n = 5)
top_restaurants
print("Most Popular Fast Food Restaurant is Mcdonlad's")

plot1 <- top_restaurants |>
    ggplot(aes(x = name, y = n)) +
    geom_bar(stat = "identity", position = "dodge") +
    labs(x = "Names of Fast Food Restaurants", y = "No. of Restaurants") +
    ggtitle("Bar graph of Most Popular Fast Food Restaurants in U.S.")
plot1


state_counts <- fast_food |>
    semi_join(top_restaurants) |> # semi_join gives the intersection of two data frames
    group_by(st) |>
    summarise(n = n()) |>
    arrange(desc(n)) |>
    tail(n=3)
population <- c(7.614, 3.017, 5.758)
state_counts |>
    mutate(n_per_capita = n / population)
print("State with least fast food restaurant count per capita is Colorado")

low_n_state <- fast_food |>
    semi_join(top_restaurants) |> # semi_join gives the intersection of two data frames
    filter(st %in% c("Co Spgs", "AK", "DC", name %in% c("McDonald's", "Burger King", "Taco Bell", "Wendy's", "Arby's"))) |> 
    group_by(name, st) |>
    summarise(n = n()) |>
    arrange(desc(n)) 
low_n_state

plot2 <- low_n_state |>
    ggplot(aes(x = st, y = n, fill = name)) + 
    geom_bar(stat = "identity", position = "dodge") + 
   xlab("Name of US State") +
   ylab("No. of restaurants") +
   labs(fill = "Restaurant") +
    ggtitle("Bar graph of States with the lowest number of restaurants")
plot2
```
